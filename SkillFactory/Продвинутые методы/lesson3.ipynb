```python
x_ = np.array([1, 2, 3, 4, 5])
y_ = np.array([2, 4, 5, 4, 5])

def mnk(x_,y_):
    # запишите далее ваш код
    A = np.array([x_, np.ones(len(x_))])
    w = np.linalg.lstsq(A.T, y_, rcond=None)[0]
    a = w[0]
    b = w[1]
    return a,b

print(mnk(x_,y_))
```
